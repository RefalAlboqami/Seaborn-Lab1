## Data Description

- Survival : 0 = No, 1 = Yes
- pclass(Ticket Class) : 1 = 1st, 2 = 2nd, 3 = 3rd
- Sex(Gender) : Male, Female
- Age : Age in years
- SibSp : Number of siblings/spouses abroad the titanic
- Parch : Number of parents/children abrod the titanic
- Ticket : Ticket Number
- Fare : Passenger fare
- Cabin : Cabin Number
- Embarked : Port of Embarkation, C = Cherbourg, Q = Queenstown, S = Southampton

# Import Seaborn

```python
import seaborn as sns
```

# Q9: Draw barplot represents survived people based on sex

```python
sns.barplot(x=df['sex'], y=df['survived'], palette=["#4361EE", "#4CC9F0"],estimator=np.mean);
```

# Q10: Draw catplot represents survived people based on embarked

```python
sns.catplot(data=df, x="survived", col="embarked",kind="count", height=4)
```

# Q11: Draw boxplot represents distribution of male and female based on age and pclass

```python
plt.figure(figsize=(10, 5))
sns.boxplot(data=df, x="sex", y="age",hue="pclass", palette =['#8de5a1','#fffea3',"#4CC9F0"])
```

# Q12: Draw heatmap represents correlations between sibsp, parch, age, fare, and survived columns

```python
data= np.random.randint(low=1, 
                         high=100, 
                         size=(10, 10)) 
annot = True
hm = sns.heatmap(data=df[["sibsp","parch","age","fare","survived"]].corr(), annot=annot) 
plt.show() 
```

# Q13: Draw factorplot represents the relation between sibsp and survived columns

```python
sns.catplot(data=df, x="sibsp", col="survived",kind="bar", height=4)
```
